```python
df1=df.drop(['area_type','availability','society'],axis=1)
```

```python
df2=df1.dropna()
```

```python
df2['BHK']=df2['size'].apply(lambda x: int(x.split(' ')[0]))
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
        
```

```python
df2[~df2['total_sqft'].apply(is_float)].head(10)
```

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens)==2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
convert_sqft_to_num("2422")
```

```python
convert_sqft_to_num("1145 - 1340")
```

```python
df3=df2.copy()
```

```python
df3["total_sqft"]=df3['total_sqft'].apply(convert_sqft_to_num)
```

```python
df3.head()
```

```python
df4=df3.copy()
```

```python
df4['price_per_sq_ft']=df4['price']*100000/df4['total_sqft']
```

```python
df4.head()
```

```python
df4.location.unique()
```

```python
df4.location=df4.location.apply(lambda x: x.strip())
```

```python
location_stats=df4.groupby('location')['location'].agg('count').sort_values(ascending=False)
```

```python
location_stats
```

```python
len(location_stats[location_stats<=10])
```

```python
len(df4.location.unique())
```

```python
location_stats_less_than_10 = location_stats[location_stats<=10]
```

```python
location_stats_less_than_10
```

```python
df4.location = df4.location.apply(lambda x: 'other' if x in location_stats_less_than_10 else x)
```

```python
len(df4.location.unique())
```

```python
df4
```

```python
df5=df4[~(df4.total_sqft/df4.BHK<300)]
```

```python
df5
```

```python
df5.price_per_sq_ft.describe()
```

```python
def remove_pps_outliers(df0):
    df_out = pd.DataFrame()
    for key, subdf in df0.groupby('location'):
        m = np.mean(subdf.price_per_sq_ft)
        st = np.std(subdf.price_per_sq_ft)
        reduced_df = subdf[(subdf.price_per_sq_ft>(m-st)) & (subdf.price_per_sq_ft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out   
```

```python
df6=remove_pps_outliers(df5)
```

```python
df6.shape
```

```python
df6
```

```python
dummies=pd.get_dummies(df6.location)
```

```python
dummies
```

```python
df7=pd.concat([df6,dummies.drop('other',axis=1)],axis=1)
```

```python
df7.drop(['size'],axis=1,inplace=True)
```

```python
X=df7.drop('price',axis=1)
```

```python
y=df7.price
```

```python
y.head()
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
X.columns
```

```python
np.zeros(len(X.columns))
```

```python
def price_predict(location,total_sqft,bath,BHK):
    loc_index = np.where(X.columns==location)[0][0]
    x = np.zeros(len(X.columns))
    x[0]=total_sqft
    x[1]=bath
    x[2]=BHK
    if loc_index >= 1:
        x[loc_index]=1
    return model.predict([x])[0]
```

```python
print('Rs.')
price_predict('Indira Nagar',3600,4,5)*100000
```
